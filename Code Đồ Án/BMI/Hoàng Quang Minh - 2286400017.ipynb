```python
list_file=gl.glob("D:/Thu thập & tiền xử lý dữ liệu/Đề Classroom/Data1_2_3_4/employees*.csv")
e1=pd.read_csv("D:/Thu thập & tiền xử lý dữ liệu/Đề Classroom/Data1_2_3_4/employees1.csv")
e2=pd.read_csv("D:/Thu thập & tiền xử lý dữ liệu/Đề Classroom/Data1_2_3_4/employees2.csv")
```

```python
print(e1)
print(e2)
```
